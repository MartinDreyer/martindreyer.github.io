```python
%pip install pandas matplotlib numpy
import pandas as pd
import json

def add_flag_columns(df, flags_dict):
    # Mapping function to fetch the flag code
    def get_flag_code(country):
        return flags_dict.get(country, {}).get("code", None)

    # Create new columns
    df['home_flag'] = df['home_team'].apply(get_flag_code)
    df['away_flag'] = df['away_team'].apply(get_flag_code)

    return df

flag_path = 'data/lookup_flag.json'
flags_dict = json.loads(open(flag_path).read())
country_dict = {country['name']: country for country in flags_dict}
df = pd.read_csv('data/4-0-tidslinje.csv')

df_with_flags = add_flag_columns(df, country_dict)

df_with_flags.to_csv('data/4-0-tidslinje-with-flags.csv', index=False)
```
